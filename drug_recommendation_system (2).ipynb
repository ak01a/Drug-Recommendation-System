# **Train test split**

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
dataset = pd.read_csv('datasets/Training.csv')
X = dataset.drop('prognosis', axis=1)
y = dataset['prognosis']

# ecoding prognonsis
le = LabelEncoder()
Y = le.fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=20)
```

# **Traning top models**

```python
!pip install scikit-learn==1.3.2
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, confusion_matrix
import numpy as np
import warnings
from sklearn.exceptions import DataConversionWarning


# Create a dictionary to store models
models = {
    'SVC': SVC(kernel='linear'),
    'RandomForest': RandomForestClassifier(n_estimators=100, random_state=42),
    'GradientBoosting': GradientBoostingClassifier(n_estimators=100, random_state=42),
    'KNeighbors': KNeighborsClassifier(n_neighbors=5, weights='uniform'),
    'MultinomialNB': MultinomialNB()
}

# Suppress warnings related to data conversion
warnings.filterwarnings("ignore", category=UserWarning, message="X does not have valid feature names")

# Loop through the models, train, test, and print results
for model_name, model in models.items():
    # Train the model
    model.fit(X_train, y_train)
    with warnings.catch_warnings():
        warnings.filterwarnings("ignore",category=UserWarning, message="X does not have valid feature names")
        model.fit(X_train, y_train)
    # Test the model
    predictions = model.predict(X_test)

    # Calculate accuracy
    accuracy = accuracy_score(y_test, predictions)
    print(f"{model_name} Accuracy: {accuracy}")

    # Calculate confusion matrix
    cm = confusion_matrix(y_test, predictions)
    print(f"{model_name} Confusion Matrix:")
    print(np.array2string(cm, separator=', '))

    print("\n" + "="*40 + "\n")

# Save all models
import pickle
for model_name, model in models.items():
    pickle.dump(model, open(f'{model_name.lower()}.pkl', 'wb'))

# Load models
svc = pickle.load(open('svc.pkl', 'rb'))
random_forest = pickle.load(open('randomforest.pkl', 'rb'))
gradient_boosting = pickle.load(open('gradientboosting.pkl', 'rb'))
k_neighbors = pickle.load(open('kneighbors.pkl', 'rb'))
multinomial_nb = pickle.load(open('multinomialnb.pkl', 'rb'))


```
